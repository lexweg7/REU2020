```python
#There's probably a smarter way to do this but it works!
events = []
events.append(hf['Event_[0]'])
events.append(hf['Event_[1]'])
events.append(hf['Event_[2]'])
events.append(hf['Event_[3]'])
events.append(hf['Event_[4]'])
events.append(hf['Event_[5]'])
events.append(hf['Event_[6]'])
events.append(hf['Event_[7]'])
events.append(hf['Event_[8]'])
events.append(hf['Event_[9]'])
```

```python
print(len(events))
```

```python
#plot 1 3d
fig = plt.figure(figsize=(20, 14))
ax = fig.add_subplot(2, 2, 1, projection='3d')
ax.set_xlim3d(left=-200.0, right=200.0) 
ax.set_ylim3d(bottom=-200.0, top=200.0) 
ax.set_zlim3d(bottom=0.0, top=1000.0)
ax.scatter([item[0] for item in events[0]], [item[1] for item in events[0]], [item[2] for item in events[0]])
ax.set_title('0')
```

```python
#plot all 3d
fig = plt.figure(figsize=(20, 14))
for i in range(10):
    ax = fig.add_subplot(4, 4, i+1, projection='3d')
    ax.set_xlim3d(left=-125.0, right=125.0) 
    ax.set_ylim3d(bottom=-125.0, top=125.0) 
    ax.set_zlim3d(bottom=0.0, top=1000.0)
    ax.scatter([item[0] for item in events[i]], [item[1] for item in events[i]], [item[2] for item in events[i]])
    ax.set_title(i) 
plt.show()
```

```python
#xy plots
plt.figure(figsize=(15, 15))
for i in range(10):
    x = str(i)
    plt.subplot(4,4,i+1)
    ax = fig.add_subplot(111)
    plt.xlim(-125, 125)
    plt.ylim(-125, 125)
    ax.set_aspect('equal', adjustable='box')
    plt.xticks()
    plt.yticks()
    plt.grid(False)
    plt.plot([item[0] for item in events[i]], [item[1] for item in events[i]], 'b.')
    plt.xlabel(x)
plt.show()
```
