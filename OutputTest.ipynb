```python
import h5py
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
hf = h5py.File('output_digi_HDF.h5', 'r')
```

```python
print(hf.keys())
```

```python
#Plot single event
x1 = [item[0] for item in hf['Event_[3]']]
y1 = [item[1] for item in hf['Event_[3]']]
```

```python
plt.plot(x1, y1, 'b.')
```

```python
events = []
for i in hf.keys():
    events.append(hf[i])
```

```python
print(len(events))
```

```python
events[0]
```

```python
#plot 1 3d
fig = plt.figure(figsize=(20, 14))
ax = fig.add_subplot(2, 2, 1, projection='3d')
ax.set_xlim3d(left=-270.0, right=270.0) 
ax.set_ylim3d(bottom=-270.0, top=270.0) 
ax.set_zlim3d(bottom=0.0, top=1000.0)
ax.scatter([item[0] for item in events[0]], [item[1] for item in events[0]], [item[2] for item in events[0]])
ax.set_title('0')
```

```python
#plot all 3d
fig = plt.figure(figsize=(20, 14))
for i in range(10):
    ax = fig.add_subplot(4, 4, i+1, projection='3d')
    ax.set_xlim3d(left=-270.0, right=270.0) 
    ax.set_ylim3d(bottom=-270.0, top=270.0) 
    ax.set_zlim3d(bottom=0.0, top=1000.0)
    ax.scatter([item[0] for item in events[i]], [item[1] for item in events[i]], [item[2] for item in events[i]])
    ax.set_title(i) 
plt.show()
```

```python
#xy plots
plt.figure(figsize=(15, 15))
for i in range(10):
    x = str(i)
    plt.subplot(4,4,i+1)
    ax = fig.add_subplot(111)
    plt.xlim(-270, 270)
    plt.ylim(-270, 270)
    ax.set_aspect('equal', adjustable='box')
    plt.xticks()
    plt.yticks()
    plt.grid(False)
    plt.plot([item[0] for item in events[i]], [item[1] for item in events[i]], 'b.')
    plt.xlabel(x)
plt.show()
```
